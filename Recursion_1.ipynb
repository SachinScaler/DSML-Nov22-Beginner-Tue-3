```python
# recursive function to calculate n factorial
def f(n):
  # base condition
  if n==0:
    return 1

  return n*f(n-1)

f(5)
```

```python
def sum_of_series(n):
  if n==0:
    return 0
  return n+sum_of_series(n-1)

sum_of_series(3) # 1 +2 +3 = 6
```

```python
# 1*2*3*...n
def multiplication(n):
  if n==0:
    return 1

  return n*multiplication(n-1)

multiplication(6)
```

```python
def power(x,y): # calculate x^ y 
  if y==0:
    return 1 
  return x*power(x,y-1) 


power(2,5)
```

```python
def power(x,y): # calculate x^ y 
  if y==1:
    return x  
  return x*power(x,y-1) 

power(2,5)
```
